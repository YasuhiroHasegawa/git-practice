```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split # trainとtest分割用
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split # trainとtest分割用
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.metrics import confusion_matrix
```

```python
wine = pd.read_csv("../dataset/wine.data",sep=",",header=None)

wifi = pd.read_csv("../dataset/wifi.txt",sep="\t",header=None)
```

```python
wine_label  = wine.iloc[:,:1] 
wine_data = wine.iloc[:, 1:10]

wifi_data = wifi.iloc[3:,:7]
wifi_label = wifi.iloc[3:,7:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data,wine_label, test_size=0.2, random_state=42)
y_train =np.reshape(y_train[0],(-1))
clf = MLPClassifier(random_state=1, max_iter=1500).fit(X_train, y_train)
print("\naccuracy ratio:")
Y_pred = clf.predict(X_test)
print(metrics.accuracy_score(y_test, Y_pred))

print("\nconfusion matrix:")
print(metrics.confusion_matrix(y_test, Y_pred))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.title("Loss Curve")
plt.plot(clf.loss_curve_)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
clf_forest = RandomForestClassifier(max_depth=10, random_state=0)
clf_forest.fit(X_train, y_train)

Y_pred = clf_forest.predict(X_test)

print("\naccuracy ratio:")

print(metrics.accuracy_score(y_test, Y_pred))

print("\nconfusion matrix:")
print(metrics.confusion_matrix(y_test, Y_pred))
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(50,50))
_ = tree.plot_tree(clf_forest.estimators_[0], feature_names=wine.columns, filled=True)
```
